Q1: Understanding Central Tendency

Data: [10, 12, 11, 15, 14, 13, 12]

Most representative value: 12 (Median/Mode)

Why: Sales are fairly consistent and 12 appears most often and lies centrally, making it a good representation of typical sales.

Q2: Mean in Real Life

Data: [12, 15, 14, 16, 18, 20, 19]

Mean = (12+15+14+16+18+20+19) ÷ 7 = 16.29 (approx)

Interpretation: The average class performance is good, with most students scoring above average.

Q3: Mode in Real Life

Data: [7, 8, 9, 8, 8, 10, 7, 9]

Mode = 8

Usefulness: Helps the store stock more shoes of the most demanded size.

Q4: Median in Real Life

Data: [$8,000, $9,500, $10,200, $11,000, $50,000]

Median = $10,200

Why median is better: The very high price ($50,000) skews the mean, while the median reflects a typical car price.

Q5: Dispersion Introduction

Data: [25, 30, 27, 35, 40]

Range = 40 − 25 = 15

Meaning: Puzzle-solving time varies by 15 minutes, showing moderate inconsistency.

Q6: Range in Action

Data: [100, 105, 98, 110, 120]

Use: Helps the farmer plan flexible packaging sizes for varying harvest weights.

Q7: Variance for Decision-Making

Q8: Standard Deviation in Context

Q9: Combining Measures

Data: [400, 420, 390, 450, 410]

Mean = (400+420+390+450+410) ÷ 5 = 414 kWh

Standard Deviation ≈ 20.6 kWh

Interpretation:
The family’s electricity usage is fairly stable with small monthly fluctuations.
